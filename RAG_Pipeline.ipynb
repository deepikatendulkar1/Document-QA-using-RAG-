```python
from langchain.document_loaders import PyMuPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load your PDF file
loader = PyMuPDFLoader("/content/drive/MyDrive/firebase.pdf")
docs = loader.load()

# Break content into 500-token chunks with overlap
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(docs)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# Load a small, fast, free embedding model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Store chunks in FAISS DB
vectordb = FAISS.from_documents(chunks, embedding_model)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
```

```python
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

rag_pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer)
```

```python
def ask_rag(query, k=3):
    # Step 1: Get relevant chunks
    docs = vectordb.similarity_search(query, k=k)
    context = "\n".join([doc.page_content for doc in docs])

    # Step 2: Format as prompt
    prompt = f"Context: {context}\n\nQuestion: {query}\nAnswer:"

    # Step 3: Generate response using Hugging Face model
    result = rag_pipeline(prompt, max_length=300)[0]['generated_text']

    return result, docs
```

```python
query = "What is Firebase and what is it used for?"
answer, sources = ask_rag(query)

print(" Answer:\n", answer)

print("\n Sources:")
for i, doc in enumerate(sources):
    print(f"\nSource {i+1}:\n{doc.page_content[:300]}")
```
